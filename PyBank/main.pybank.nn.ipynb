```python
from pathlib import Path
import csv

csvpath = Path('C:/Repositories/Homework2_Python/python-homework/PyBank/homework2_pybank_budget.csv')

line_num = 0
pl = []
month_list = []

with open(csvpath, 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    header = next(csvreader)
    line_num += 1
    print(f"{header} <--- HEADER!\n")
    for row in csvreader:
        print(row)
        month = row[0]
        monthly_pl = int(row[1])
        month_list.append(month)
        pl.append(monthly_pl)
        
count_months = 0
total_pl = 0
avg_pl = 0
max_profit = 0
min_profit = 0
previous_month = pl[0]
total_change = 0
month_min = 0
month_max = 0
    
for monthly_pl in pl:
    count_months += 1
    total_pl += monthly_pl
    
    total_change += monthly_pl - previous_month
    previous_month = monthly_pl
    
    if min_profit == 0:
        min_profit = monthly_pl
    elif monthly_pl > max_profit:
        max_profit = monthly_pl
        month_max = count_months
    elif monthly_pl < min_profit:
        min_profit = monthly_pl
        month_min = count_months
        
avg_pl = round(total_pl / count_months, 2)
avg_change = round(total_change / (count_months - 1), 2)

# outputs to terminal

print(f"\nFinancial Analysis (to Terminal):")
print("---------------------------------")
print(f"The max profit is: {month_list[int(month_max) - 1]} $",format(max_profit, ",.2f"))
print(f"The min profit is: {month_list[int(month_min) - 1]} $",format(min_profit, ",.2f"))
print(f"The avg profit/loss is: $",format(avg_pl, ",.2f"))
print(f"The avg change is: $",format(avg_change, ",.2f"))
print(f"The total profit/loss for the period is: $",format(total_pl, ",.2f"))
print(f"The total number of months in the period is: {count_months}")

# outputs to csv

output_max_profit = (f"The max profit is: {month_list[int(month_max) - 1]} ${max_profit}")
output_min_profit = (f"The min profit is: {month_list[int(month_min) - 1]} ${min_profit}")
output_avg_pl = (f"The avg profit/loss is: ${avg_pl}")
output_avg_change = (f"The avg change is: ${avg_change}")
output_total_pl = (f"The total profit/loss for the period is: ${total_pl}")
output_count_months = (f"The total number of months in the period is: {count_months}")

header_1 = ["Financial Analysis:"]
line_1 = [output_max_profit]
line_2 = [output_min_profit]
line_3 = [output_avg_pl]
line_4 = [output_avg_change]
line_5 = [output_total_pl]
line_6 = [output_count_months]

output_path = Path('main.pybank.output.nn.csv')
                  
with open(output_path, 'w') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')
    csvwriter.writerow(header_1)
    csvwriter.writerow(line_1)
    csvwriter.writerow(line_2)
    csvwriter.writerow(line_3)
    csvwriter.writerow(line_4)
    csvwriter.writerow(line_5)
    csvwriter.writerow(line_6)
```
